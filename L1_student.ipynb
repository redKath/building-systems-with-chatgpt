```python

import os
import openai
import tiktoken
from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv()) # read local .env file

openai.api_key  = os.environ['OPENAI_API_KEY']
```

```python
client = openai.OpenAI()

def get_completion(prompt, model="gpt-3.5-turbo"):
    messages = [{"role": "user", "content": prompt}]
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=0
    )
    return response.choices[0].message.content
```

```python
response = get_completion("What is the capital of South Korea?")
print(response)
```

```python
response = get_completion("Take the letters in guacamole \
and reverse them")
print(response)
```

```python
response = get_completion("""Take the letters in \
g-u-a-c-a-m-o-l-e and reverse them""")
response
```

```python
client = openai.OpenAI()

def get_completion_from_messages(messages, 
                                 model="gpt-3.5-turbo", 
                                 temperature=0, 
                                 max_tokens=500):
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=temperature, # this is the degree of randomness of the model's output
        max_tokens=max_tokens, # the maximum number of tokens the model can ouptut 
    )
    return response.choices[0].message.content
```

```python
messages =  [  
{'role':'system', 
 'content':"""You are an assistant who\
 responds in the style of Carl Jung."""},    
{'role':'user', 
 'content':"""write me a very short poem\
 about a blueberry cheese cake."""},  
] 
response = get_completion_from_messages(messages, temperature=1)
print(response)
```

```python
# length
messages =  [  
{'role':'system',
 'content':'All your responses must be \
one sentence long.'},    
{'role':'user',
 'content':'write me a story about a sad blueberry cheese cake.'},  
] 
response = get_completion_from_messages(messages, temperature =1)
print(response)
```

```python
# combined
messages =  [  
{'role':'system',
 'content':"""You are an assistant who \
responds in the style of Carl Jung. \
All your responses must be one sentence long."""},    
{'role':'user',
 'content':"""write me a story about a happy blueberry cheese cake."""},
] 
response = get_completion_from_messages(messages, 
                                        temperature =1)
print(response)
```

```python
client = openai.OpenAI()

def get_completion_and_token_count(messages, 
                                   model="gpt-3.5-turbo", 
                                   temperature=0, 
                                   max_tokens=500):
    
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=temperature, 
        max_tokens=max_tokens,
    )
    
    content = response.choices[0].message.content
    
    token_dict = {
'prompt_tokens':response.usage.prompt_tokens,
'completion_tokens':response.usage.completion_tokens,
'total_tokens':response.usage.total_tokens,
    }

    return content, token_dict
```

```python
messages = [
{'role':'system', 
 'content':"""You are an assistant who responds in the style of Carl Jung."""},    
{'role':'user',
 'content':"""write me a very short poem about a yearning blueberry cheese cake"""},  
] 
response, token_dict = get_completion_and_token_count(messages)
```
